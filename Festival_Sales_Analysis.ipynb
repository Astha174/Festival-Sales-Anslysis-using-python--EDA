```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt #for data visualization
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv('Festival Sales Data.csv' ,encoding='unicode_escape')
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'] ,axis=1,inplace=True)
```

```python
df.info()
```

```python
df.shape
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df['Amount']=df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

```python
df.rename(columns={'Marital_Status':'Marriage'})
```

```python
df[['Age','Orders','Amount']].describe()
```

**Gender**

```python
df.columns
```

```python
sns.countplot(x='Gender',data=df)
```

```python
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index=False )['Amount'].sum().sort_values(by='Amount' ,ascending=False)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False )['Amount'].sum().sort_values(by='Amount' ,ascending=False)
sns.barplot(  x='Gender', y='Amount',data=sales_gen)
```

From graphs above we conclude that most of the shopping has been done by females and spending of females is more.

**AGE**

```python
ax=sns.countplot(data=df ,   x='Age Group',   hue='Gender')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_age=df.groupby(['Age Group'],as_index=False )['Amount'].sum().sort_values(by='Amount' ,ascending=False)
sns.barplot(  x='Age Group', y='Amount',data=sales_age)
```

```python
sales_state=df.groupby(['State'],as_index=False )['Orders'].sum().sort_values(by='Orders' ,ascending=False).head(10)
sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(  data=sales_state ,x='State', y='Orders')
```

```python
sales_state=df.groupby(['State'],as_index=False )['Amount'].sum().sort_values(by='Amount' ,ascending=False).head(10)
sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(  data=sales_state ,x='State', y='Amount')
```

We see most orders are from UP,Maharastra and Karnataka

**Marital Status**

```python
ax=sns.countplot(data=df, x='Marital_Status')
sns.set(rc={'figure.figsize':(3,5)})
for bars in ax.containers:
 ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Marital_Status','Gender'],as_index=False )['Amount'].sum().sort_values(by='Amount' ,ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(  data=sales_state ,x='Marital_Status', y='Amount',hue='Gender')
```

Most of shopping is done by Married women and Expenditure rate of married women is more

**Occupation**

```python
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(data=df , x='Occupation')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Occupation','Gender'],as_index=False )['Amount'].sum().sort_values(by='Amount' ,ascending=False)
sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(  data=sales_state ,x='Occupation', y='Amount',hue='Gender')
```

We can see most of the buyers are from IT sector ,Aviation and healthcare

**Product** **Category**

```python
sns.set(rc={'figure.figsize':(24,5)})
ax=sns.countplot(data=df , x='Product_Category')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Product_Category','Gender'],as_index=False )['Amount'].sum().sort_values(by='Product_Category' ,ascending=False)
sns.set(rc={'figure.figsize':(25,3)})
sns.barplot(  data=sales_state ,x='Product_Category', y='Amount',hue='Gender')
```

```python
sales_state=df.groupby(['Product_Category','Orders'],as_index=False )['Orders'].sum().sort_values(by='Product_Category' ,ascending=False)
sns.set(rc={'figure.figsize':(25,3)})
sns.barplot(  data=sales_state ,x='Product_Category', y='Orders')
```

Most products sold are Food ,Electronics and gadget and clothing

```python
sales_state=df.groupby(['Product_ID'],as_index=False )['Orders'].sum().sort_values(by='Product_ID' ,ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,3)})
sns.barplot(  data=sales_state ,x='Product_ID', y='Orders')
```

**Conclusion**
We conclude that married women between the age group 26-35 years from Up ,Maharastra and Karnataka working in IT,Healthcare and AViation are more likely to buy Food,Clothing and  Electronics.
